## Hands on Lab


##### Step 1: Install and Import Required Libraries


```python
# Install the necessary libraries
"""!pip install pandas
!pip install matplotlib
!pip install seaborn"""

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2: Load the Dataset


```python
# Load the dataset from the given URL
file_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_url)

# Display the first few rows to understand the structure of the dataset
df.head()
```

<h3>Step 3: Analyze and Visualize Compensation Distribution</h3>


**Task**: Plot the distribution and histogram for `ConvertedCompYearly` to examine the spread of yearly compensation among respondents.


```python
## Write your code here
plt.figure(figsize=(10, 6))
plt.hist(df['ConvertedCompYearly'], bins = 20, color = 'green')
plt.title('Distribution of Yearly converted compensation', fontsize = 16)
plt.xlabel('Yearly converted compensation', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

<h3>Step 4: Calculate Median Compensation for Full-Time Employees</h3>


**Task**: Filter the data to calculate the median compensation for respondents whose employment status is "Employed, full-time."


```python
## Write your code here
full_time_employed = df[df['Employment'] == 'Employed, full-time']
median_compensation = full_time_employed['ConvertedCompYearly'].median()
print(f"The median compensation is: {median_compensation}")
```

<h3>Step 5: Analyzing Compensation Range and Distribution by Country</h3>


Explore the range of compensation in the ConvertedCompYearly column by analyzing differences across countries. Use box plots to compare the compensation distributions for each country to identify variations and anomalies within each region, providing insights into global compensation trends.



```python
# Remove rows with missing or invalid compensation data (e.g., zero, negative, or NaN values)
df_clean = df.dropna(subset=['ConvertedCompYearly'])
df_clean = df_clean[df_clean['ConvertedCompYearly'] > 0]
#Create a box plot to compare the distribution of compensation across countries
plt.figure(figsize=(18, 8))
sns.boxplot(x='Country', y='ConvertedCompYearly', data=df_clean)

# Set labels and title for better understanding
plt.title('Distribution of Yearly Compensation by Country', fontsize=16)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Converted Compensation (Yearly)', fontsize=12)
plt.xticks(rotation=90)  # Rotate x-axis labels to avoid overlap
plt.show()
```

<h3>Step 6: Removing Outliers from the Dataset</h3>


**Task**: Create a new DataFrame by removing outliers from the `ConvertedCompYearly` column to get a refined dataset for correlation analysis.


```python
## Write your code here
#We will use the IQR method to find these values
Q1 = df['ConvertedCompYearly'].quantile(0.25)
Q3 = df['ConvertedCompYearly'].quantile(0.75)
IQR = Q3-Q1

#Calculating upper and lower limits using the IQR, Q1 and Q3
upper_limit= Q3+1.5*IQR
lower_limit = Q1-1.5*IQR

upper_bound = df['ConvertedCompYearly'] >= upper_limit
lower_bound = df['ConvertedCompYearly'] <= lower_limit

# Combine the masks (logical OR) for rows outside the range
out_of_range_bound = upper_bound | lower_bound

# Drop rows where the condition is True
new_df = df[~out_of_range_bound]
```

<h3>Step 7: Finding Correlations Between Key Variables</h3>


**Task**: Calculate correlations between `ConvertedCompYearly`, `WorkExp`, and `JobSatPoints_1`. Visualize these correlations with a heatmap.


```python
## Write your code here
new_df = new_df.dropna(subset=['ConvertedCompYearly', 'WorkExp', 'JobSatPoints_1'], axis=0)
df_corr = new_df[['ConvertedCompYearly', 'WorkExp', 'JobSatPoints_1']].corr()
```

```python
sns.heatmap(df_corr,cmap ="YlOrRd",fmt='.2f', annot=True)
plt.show()
```

<h3>Step 8: Scatter Plot for Correlations</h3>


**Task**: Create scatter plots to examine specific correlations between `ConvertedCompYearly` and `WorkExp`, as well as between `ConvertedCompYearly` and `JobSatPoints_1`.


```python
# Create a figure with two subplots
plt.figure(figsize=(14, 6))

# Scatter plot between ConvertedCompYearly and WorkExp
plt.subplot(1, 2, 1)  # (rows, columns, position)
sns.scatterplot(y='WorkExp', x='ConvertedCompYearly', data=new_df)
plt.title('Scatter Plot: Work Experience vs Compensation')
plt.xlabel('Work Experience (Years)')
plt.ylabel('Converted Compensation (Yearly)')

# Scatter plot between ConvertedCompYearly and JobSatPoints_1
plt.subplot(1, 2, 2)  # (rows, columns, position)
sns.scatterplot(y='JobSatPoints_1', x='ConvertedCompYearly', data=new_df)
plt.title('Scatter Plot: Job Satisfaction vs Compensation')
plt.xlabel('Job Satisfaction Points')
plt.ylabel('Converted Compensation (Yearly)')
plt.tight_layout()

plt.show()
```
